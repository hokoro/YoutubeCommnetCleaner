```python
import pandas as pd 
import matplotlib.pyplot as plt 
from konlpy.tag import Okt
```

```python
df = pd.read_csv('/content/data.csv',encoding = 'utf-8',index_col = 0)
```

```python
df
```

```python
okt = Okt()
print(okt.morphs(df['Sentence'][0]))

voca_dic = dict()
```

```python
for i in range(df.shape[0]):
  vocab = okt.morphs(df['Sentence'][i])
  for voca in vocab:
    if voca in voca_dic:
      voca_dic[voca] += 1
    else:
      voca_dic[voca] = 1

```

```python
voca_dic
```

```python
voca_dic
```

```python
voca_df = pd.DataFrame(columns = ['voca','count'])
```

```python
voca_df
```

```python
voca_df['voca'] = voca_dic.keys()
```

```python
voca_df['count'] = voca_dic.values()
```

```python
voca_df
```

```python
voca_df = voca_df.sort_values(by = 'count',ascending = False)
```
